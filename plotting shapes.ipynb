# Plotting a shape file in python using the shapefile lib
This example used data from [Datapunt Amsterdam](https://data.amsterdam.nl/#?dte=catalogus%2Fapi%2F3%2Faction%2Fpackage_show%3Fid%3D7c3e1e75-b7ea-4524-bb5c-448230088d28&dtfs=T&dsf=groups::geografie:res_format::SHP&mpb=topografie&mpz=11&mpv=52.3731081:4.8932945)

```python
import shapefile
from shapely.geometry import Polygon, Point, MultiPolygon
import matplotlib.pyplot as plt

sf = shapefile.Reader("shp/bc2010zw_region.shp")
shapes = sf.shapes()

plt.figure()
for shape in shapes:
    x,y = Polygon(shape.points).exterior.xy
    plt.plot(x,y)
plt.show()       
    
```
